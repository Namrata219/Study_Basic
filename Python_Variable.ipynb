
A variable can have a short name (like x and y) or a more descriptive name (age, carname, total_volume). Rules for Python variables:
A variable name must start with a letter or the underscore character
A variable name cannot start with a number
A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
Variable names are case-sensitive (age, Age and AGE are three different variables)

```python
#Multi Words Variable Names
```

```python
#Many Values to Multiple Variables
```

Python allows you to assign values to multiple variables in one line.

Note: Make sure the number of variables matches the number of values, or else you will get an error.

```python
x,y,z="Orange","Apple", "Cherry"
print(x)
print(y)
print(z)
```

```python
#One Value to Multiple Variables
```

We can assign the same value to multiple variables in one line:

```python
x=y=z= "Orange"
print(x)
print(y)
print(z)
```

```python
#Unpack a Collection
```

When we have a collection of values in a list, tuple etc. 
Python allows you to extract the values into variables. This is called unpacking

```python
fruits=["Orange","Apple","Cherry"]
x,y,z=fruits
print(x,y,z)
```

```python
fruits=("Orange","Apple","Cherry")
x,y,z=fruits
print(x)
print(y)
print(z)
```

# Global Variables

Variables that are created outside of a function (as in all of the examples above) are known as global variables.

Global variables can be used by everyone, both inside of functions and outside.

```python
x="simple"
def myfunct():
    print("Python is " + x)
myfunct()
```

If you create a variable with the same name inside a function,this variable will be local,and can only be used inside the function. 
The global variable with the same name will remain as it was, global and with the original value.

```python
x="simple"
def myfunct():
    x='userfriendly'
    print("Python is " + x)
myfunct()
```

```python
#to print global variable we need to command print outside the function
```

```python
x="simple"
def myfunct():
    x='userfriendly'
    print("Python is " + x)
myfunct()

print("Python is " + x)
```

Global keyword
Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function.

To create a global variable inside a function, you can use the global keyword.

```python
def myfunct():
    global x
    x='userfriendly'
    print("Python is " + x)
myfunct()
```

```python
def myfunct():
    global x
    x='userfriendly'
    
myfunct()
print("Python is " + x)
```

```python
#To change the value of a global variable inside a function, refer to the variable by using the global keyword:
```

```python
x="simple"
def myfunct():
    global x
    x= "usefriendly"
myfunct()

print("Python is " + x)
```
